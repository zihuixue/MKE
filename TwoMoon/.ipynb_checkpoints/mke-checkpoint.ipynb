```python
import numpy as np
import copy
import random
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_moons, make_circles, make_classification
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC

import torch
from torch.autograd import Variable
from torch.nn import functional as F
```

```python
class TwoLayerNN(torch.nn.Module):
    def __init__(self, input_dim, hidden_dim, dropout_rate, seed):
        super(TwoLayerNN, self).__init__()
        torch.manual_seed(seed)
        self.fc1 = torch.nn.Linear(input_dim, hidden_dim)
        self.fc2 = torch.nn.Linear(hidden_dim, hidden_dim)
        self.fc3 = torch.nn.Linear(hidden_dim, 2)
        self.relu = torch.nn.ReLU()
        self.softmax = torch.nn.Softmax(dim=1)
        self.dropout = torch.nn.Dropout(dropout_rate)
        
    def forward(self, x, beta):
        y = self.dropout(self.relu((self.fc1(x))))
        y = y + beta * abs(y).mean() * torch.randn(y.shape)
        y = self.dropout(self.relu((self.fc2(y))))
        y = self.softmax(self.fc3(y))
        return y
```

```python
### Generate Data
h = 0.02  # step size in the mesh

datasets = make_moons(n_samples=500, noise=0.1, random_state=42)

# preprocess dataset, split into training and test part
X, y = datasets
X = StandardScaler().fit_transform(X)

# split into labeled and unlabeled data
X_train, X_t, y_train, y_t = train_test_split(X, y, test_size=.4, random_state=42)
X_l, X_u, y_l, y_u = train_test_split(X_train, y_train, test_size=.9, random_state=42)  # 500, 0.4, 0.7

# plot the dataset first
print('Data visualization')
plt.figure(figsize=(8,5))
ax = plt.subplot(1,1,1)
cm = plt.cm.RdBu
cm_bright = ListedColormap(['#FF0000', '#0000FF'])
# plot the labeled points
plt.scatter(X_l[:, 0], X_l[:, 1], c=y_l, cmap=cm_bright, edgecolors='k', s=150)
# plot the unlabeled points
plt.scatter(X_u[:, 0], X_u[:, 1], c=y_u, cmap=cm_bright, alpha=0.15)
plt.xlabel('Xa')
plt.ylabel('Xb')
ax.set_xticks(())
ax.set_yticks(())
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
```

```python
### Some functions

def plot_decision_boundary(pred_func,X,y,d, score=None, contour=True):
    plt.figure(figsize=(8,5))
    ax = plt.subplot(1,1,1)
    # Set min and max values and give it some padding
    x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
    y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
    h = 0.01
    # Generate a grid of points with distance h between them
    xx,yy=np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    # Predict the function value for the whole gid
    if d == 1:
        Z = pred_func(np.c_[xx.ravel()])
    if d == 2:
        Z = pred_func(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    # Plot the contour and training examples
    if contour:
        plt.contourf(xx, yy, Z, cmap=cm_bright, alpha=0.2)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cm_bright, alpha=0.4)
    if score:
        ax.text(xx.max() - .3, yy.min() + .3, ('%.2f' % score), size=30, family='Times New Roman', horizontalalignment='right')
    plt.xlabel('Xa')
    plt.ylabel('Xb')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.set_xticks(())
    ax.set_yticks(())

    
def predict(x, m):
    m.eval()
    # Convert into numpy element to tensor
    x = torch.from_numpy(x).type(torch.FloatTensor)
    # Predict and return ans
    output = m(x, 0)
    _, predicted = torch.max(output, 1)
    return predicted.numpy()

    
def train(seed, mod, x_data, y_data, x_test, y_test, hidden_dim, epochs, eval_epochs, alpha, beta, dp, save_model=False, plot=True):
    model = TwoLayerNN(mod, hidden_dim, dp, seed)
    criterion = torch.nn.CrossEntropyLoss()
    optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
    val_acc = []
    for epoch in range(epochs):
        model.train()
        optimizer.zero_grad()
        # Forward pass
        x = x_data + alpha * abs(x_data).mean() * torch.randn(x_data.shape)
        y_pred = model(x, beta)
        # Compute Loss
        loss = criterion(y_pred, y_data)
        # Backward pass
        loss.backward()
        optimizer.step()
    
        model.eval()
        if epoch % eval_epochs == eval_epochs-1:
            with torch.no_grad():
                outputs = model(x_test, 0)
            _, predicted = torch.max(outputs, 1)
            c = (predicted == y_test).squeeze().sum()
            acc = c.item() /len(y_test) * 100
            val_acc.append(acc)
            
        if plot and (epoch % epochs == epochs-1):
            plt.plot(val_acc)
            plt.xlabel('epoch')
            plt.ylabel('test acc')
            print('Epoch: %d \t loss: %.2f \t test acc: %.2f' %(epoch, loss, acc))
    
    if save_model:
        return model
    else:
        return acc
```

```python
### Step1: train a teacher model

x_train = torch.FloatTensor(X_l[:, 0]).view(-1,1)
y_train = torch.LongTensor(y_l)
x_test = torch.FloatTensor(X_t[:, 0]).view(-1,1)
y_test = torch.LongTensor(y_t)

# train a teacher model based on labeled unimodal data (Xl)
teacher = train(39, 1, x_train, y_train, x_test, y_test, 32, 1000, 10, 0, 0, 0, save_model=True, plot=False)

# visualize teacher's predictions on labeled data
print('Teacher predictions on labeled and unlabeled data')
pred = predict(X_t[:,0].reshape(-1,1), teacher)
s_score = np.count_nonzero(pred==y_t) / len(y_t)
plot_decision_boundary(lambda x:predict(x, teacher), X_l, y_l, 1, s_score*100)
plt.scatter(X_l[:, 0], X_l[:, 1], c=y_l, cmap=cm_bright, edgecolors='k', s=200)

# visualize teacher's predictions on unlabeled data
y_pl = predict(X_u[:,0].reshape(-1,1), teacher)
plot_decision_boundary(lambda x:predict(x, teacher), X_u, y_pl, 1, contour=False)
```

```python
### Step2: MKE, train a multimodal student that learns from the unimodal teacher

x_train = torch.FloatTensor(X_u[:, 0]).view(-1,1)
y_pl = predict(X_u[:,0].reshape(-1,1), teacher)
y_train = torch.LongTensor(y_pl)
y_true = torch.LongTensor(y_u)
x_test = torch.FloatTensor(X_t[:, 0]).view(-1,1)
y_test = torch.LongTensor(y_t)

x_train_mm = torch.FloatTensor(X_u)
x_test_mm = torch.FloatTensor(X_t)
    
log1 = []
log2 = []
log3 = []
epochs = 2000

# three kinds of regularization, alpha / beta / dropout rate
alpha = 1
beta = 0
rate = 0

random.seed(42)
n_runs = 10
for j in range(n_runs):  
    seed = random.randint(0, 100)
    
    um = train(seed, 1, x_train, y_train, x_test, y_test, 32, epochs, epochs, alpha, beta, rate, plot=False)
    mm = train(seed, 2, x_train_mm, y_train, x_test_mm, y_test, 16, epochs, epochs, alpha, beta, rate, plot=False)
    mm_sup = train(seed, 2, x_train_mm, y_true, x_test_mm, y_test, 16, epochs, epochs, alpha, beta, rate, plot=False)
    print(f'{j}th run | um acc {um} | mm acc {mm} | mm (sup) acc {mm_sup}')
    log1.append(um)
    log2.append(mm)
    log3.append(mm_sup)

print()
print(f'Results over {n_runs} runs:')
print(f'unimodal student {np.mean(log1)}')  # unimodal baseline
print(f'multimodal student (ours) {np.mean(log2)}')  
print(f'multimodal student (supervised) {np.mean(log3)}')  # upper bound of our multimodal student
```

```python
# visualization: compare the multimodal student's predictions with the unimodal student's predictions

um_student = train(45, 1, x_train, y_train, x_test, y_test, 32, 1000, 1000, 0, 0, 0, save_model=True, plot=False)
mm_student = train(45, 2, x_train_mm, y_train, x_test_mm, y_test, 16, 1000, 1000, 1, 0, 0, save_model=True, plot=False)  # 845, 94

print('Unimodal student predictions on test data')
pred = predict(X_t[:,0].reshape(-1,1), um_student)
s_score = np.count_nonzero(pred==y_t) / len(y_t)
plot_decision_boundary(lambda x:predict(x, um_student), X_u, y_u, 1, s_score*100)
```

```python
print('Multimodal student (ours) predictions on test data')
pred = predict(X_t, mm_student)
s_score = np.count_nonzero(pred==y_t) / len(y_t)
plot_decision_boundary(lambda x:predict(x, mm_student), X_u, y_u, 2, s_score*100) # our MM student gives much better predictions
```
